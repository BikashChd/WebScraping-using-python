```python
from bs4 import BeautifulSoup
import requests
url=BeautifulSoup("https://www.worldometers.info/coronavirus/","html.parser")
# print(url)
soup=requests.get(url)
soup=soup.text
# soup
```

```python
code=BeautifulSoup(soup,"lxml")
code.table
```

```python
data=[]
tags=code.find_all('tr')
# tags
for tag in tags:
    x=tag.text
    data.append(x)
data
```

```python
datas=[]
for i in data:
    datas.append(i.split('\n')[1:-1])
datas
```

```python
cleaned_data=[i for i in  datas if i[0]!='']
cleaned_data
```

```python
import csv
file=open('covid_data.csv','w')
x=csv.writer(file)
for i in cleaned_data:
    x.writerow(i)

file.close()
```

```python
import pandas as pd 
df=pd.read_csv('covid_data.csv')
df
```

```python
country=list(df['Country,Other'])[0:10]
total_cases=list(df['TotalCases'])[0:10]
total_cases

cleaned_cases=[] 
for cases in total_cases:
    x=int(cases.replace(',',''))
    cleaned_cases.append(x)

cleaned_cases

```
